#### Using the '=' operator vs copy() method

```python
df_02 = df.copy() #The copy() method for a DataFrame creates a true copy of the DataFrame.
df_03 = df #There is only one object (the DataFrame), and both df and df_03 refer to it and does not create a copy of our DataFrame.
```

#### Slicing rows in data frame

```python
df[0:3] # Selects the rows from 0 to 2.
df[:5] # Selects the rows till row 4.
df[-1:] # Selects the last row.
```

#### Slicing Subsets of Rows and Columns
##### 1) loc is primarily label based indexing. Integers may be used but they are interpreted as a label.
##### 2) iloc is primarily integer based indexing.

```python
df.iloc[:,0:2] #Selects all rows and first 2 columns in the dataframe.
```

#### Manipulation of Data using Conditions

```python
df = df.astype({"customer_id": int}) #converting datatype of customer_id column to integer
df[(df.customer_id >= 20201705) & (df.customer_id <= 20202195)] #filtering data based on customer_id
```

```python
df.query('(customer_id >= 20201705) & (customer_id <= 20202195)') #query() method that allows selection using an expression.
```
